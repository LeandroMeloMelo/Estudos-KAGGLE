```
#importado as bibliotecas necessarias
import pandas as pd
from folium import plugins
import numpy as np
```

```
#variavel arquivo recebendo o data set
arquivo = ('olist_geolocation_dataset.csv')
```

```
#lendo a variavel arquivo
df = pd.read_csv(arquivo)
```

```
#exibindo o dataset
df.head()
```

```
df.shape
```

```
# Informações da base de dados:
df.info()
```

```
# Extraindo as coordenadas de 1000163 registros:
coordenadas=[]
for lat,lng in zip(df.geolocation_lat.values[:1000163],df.geolocation_lng.values[:1000163]):
  coordenadas.append([lat,lng])
```

```
# Extraindo as coordenadas de 1000163 registros:
coordenadas=[]
for lat,lng in zip(df.geolocation_lat.values[:1000163],df.geolocation_lng.values[:1000163]):
  coordenadas.append([lat,lng])
```

```
# Renderizando o mapa com algumas personalizações:
mapa = folium.Map(location=[-15.788497,-47.879873],zoom_start=5,tiles='Stamen Toner')
```

```
# Adicionando os registros no mapa de calor:
mapa.add_child(plugins.HeatMap(coordenadas))        
mapa
```

```
#Visualizando os dados mais distribuidos no mapa
df.geolocation_state.unique()
```

```
# Selecionando uma amostra aleatória de 3%:
df2 = df.sample(frac = 0.03)
```

```
# Contagem dos registros:
df2.geolocation_state.value_counts()
```

```
#Vamos repetir todos os passos anteriores e ver 
#se a visualização ficou mais adequada.
coordenadas=[]
for lat,lng in zip(df2.geolocation_lat.values,df2.geolocation_lng.values):
  coordenadas.append([lat,lng])
```

```
mapa = folium.Map(location=[-15.788497,-47.879873],zoom_start=5,tiles='Stamen Toner')
```

```
mapa.add_child(plugins.HeatMap(coordenadas))        
mapa
```

```
#Persistindo o mapa em disco
mapa.save('mapa-consumidor.html')
```
